```python
import git
import os

# Clone the Flask and Django repositories
repos = {
    "flask": "https://github.com/pallets/flask.git",
    "django": "https://github.com/django/django.git"
}
repo_paths = {name: f"./{name}_repo" for name in repos}

for name, url in repos.items():
    if not os.path.exists(repo_paths[name]):
        git.Repo.clone_from(url, repo_paths[name])
```

```python
import pyref

def extract_refactoring_data(repo_path):
    refactorings = []
    for root, dirs, files in os.walk(repo_path):
        for file in files:
            if file.endswith(".py"):
                file_path = os.path.join(root, file)
                try:
                    refactoring_info = pyref.analyze_file(file_path)
                    refactorings.extend(refactoring_info)
                except Exception as e:
                    print(f"Error processing file {file_path}: {e}")
    return refactorings

flask_refactorings = extract_refactoring_data(repo_paths["flask"])
django_refactorings = extract_refactoring_data(repo_paths["django"])

print(f"Extracted {len(flask_refactorings)} refactorings from Flask.")
print(f"Extracted {len(django_refactorings)} refactorings from Django.")
```
